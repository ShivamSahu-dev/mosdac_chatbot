```python

import pandas as pd
from mosdac_utils import fetch_satellite_sensors_data, process_satellite_sensors_data
```

```python
satellite_df = pd.read_csv(r"C:\Users\WELCOME\Desktop\mosdac\raw_data\satellite_ids.csv")
```

```python
all_dataframes = []

for idx, row in satellite_df.iterrows():
    satellite_id = row["id"]
    satellite_name = row["name"]
    
    sensor_data = fetch_satellite_sensors_data(satellite_id)
    processed_df = process_satellite_sensors_data(satellite_name, satellite_id, sensor_data)
    
    if processed_df is not None:
        all_dataframes.append(processed_df)

final_df = pd.concat(all_dataframes, ignore_index=True)
```
